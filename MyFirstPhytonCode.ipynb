```python
# Matriz de Correlacion
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Let¬¥s build the values lists

value_x_list = [1,2,2,3,4,4,5,6] # [kg]
value_y_list = [2,3,4,4,4,6,5,7] # [m]

data_dict = {"X":value_x_list, "Y":value_y_list} #,
              #"BMI":value_bmi_list}

df = pd.DataFrame(data_dict)
print(df)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
#sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
#plt.title('Correlation Matrix Heatmap')
#plt.show()

y_mean = df['Y'].mean()
x_mean = df['X'].mean()

print('y_mean',y_mean)
print('x_mean',x_mean)

df['X*Y']= df['X']*df['Y']
df['X^2']= df['X']**2
df['Y^2']= df['Y']**2

print(df)

def pendiente_intercepto(df, x_col, y_col):
    n = len(df)
    sum_x = df[x_col].sum()
    sum_y = df[y_col].sum()
    sum_xy = (df[x_col] * df[y_col]).sum()
    sum_x2 = (df[x_col] ** 2).sum()
    #  pendiente m
    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)

    #  intercepto b
    b = (sum_y - m * sum_x) / n

    return m, b

m,b = pendiente_intercepto(df,'X','Y')
print(m,b)

print('m',m)
print('b',b)


print(f"La ecuaci√≥n de regresi√≥n lineal es: y = {m}x + {b}")

def regresion_lineal(x):
    return m * x + b
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['X'], df['Y'], color='blue', label='Puntos')

plt.plot(df['X'], m * df['X'] + b, color='red', label=f'Recta: y = {m:.2f}x + {b:.2f}')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Grafico de datos y recta')
plt.legend()

# gr√°fico
plt.show()
```

```python
#Errores definidos

df['y_pred'] = m * df['X'] + b

df['error'] = df['Y'] - df['y_pred']

df['error_cuadrado']=df['error']**2

SR = df['error_cuadrado'].sum()

df['error absoluto']=df['error_cuadrado'].abs()

#return y_pred, error, error_cuadrado, sr

#print('y_pred',y_pred)
#print('error',error)
#print('error_cuadrado',error_cuadrado)
#print('sr',sr)

print(f"El valor de SR (Residuos al cuadrado): {SR}")
```
